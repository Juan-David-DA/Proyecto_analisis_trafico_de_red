```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar los datos
df = pd.read_csv('Train_data.csv')
```

```python
#2. Exploración de los datos

df.head()

df.columns

#df.describe()
```

```python
# 3. Análisis Inicial: ¿Qué tipos de tráfico tenemos?
print("Distribución de Tráfico:")
print(df['class'].value_counts(normalize=True) * 100)
```

```python
# 4. Limpieza: Filtrar solo ataques para ver protocolos más peligrosos
ataques = df[df['class'] == 'anomaly']

ataques
```

```python
# 5. Agrupación: ¿Qué protocolos usan los atacantes?
protocol_stats = ataques.groupby('protocol_type')['src_bytes'].sum().reset_index()

protocol_stats
```

```python
# 6. Exportar para Power BI
# Creamos una columna de 'Riesgo' basada en la duración de la conexión
df['nivel_riesgo'] = df['duration'].apply(lambda x: 'Alto' if x > 1000 else 'Bajo')

df.to_csv('datos_limpios_seguridad.csv', index=False)
```
